```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the data
url = "https://corgis-edu.github.io/corgis/datasets/csv/smoking/smoking.csv"
df = pd.read_csv(url)

# 2. Save the file
df.to_csv("global_smoking_data.csv", index=False)
print("File saved successfully as 'global_smoking_data.csv'")

# 3. Process the data
df_clean = df.rename(columns={
    "Data.Daily cigarettes": "daily_cigs",
    "Data.Percentage.Total": "total_percentage",
    "Country": "country",
    "Year": "year"
})

us_data = df_clean[df_clean['country'] == 'United States']

# 4. Visualize
plt.figure(figsize=(10, 6))
plt.plot(us_data['year'], us_data['total_percentage'], marker='o', color='red', linestyle='-')

# Adding labels and title
plt.title('Smoking Prevalence in the USA (1980-2012)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage of Population Who Smoke', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
